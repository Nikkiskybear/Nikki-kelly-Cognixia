```python
from datetime import datetime
#get the name then making sure all the letters are lower cases and each part of the name is capiltized 
name = input("Please enter your name: ")
name = name.lower()
name = name.title()
#getting the age and birth year of the person
age = int(input("please enter your age: "))
birthYear = datetime.now().year - age

#creating a list of all the parts of the name
nameConcatList = name.split()
#creating a concat of the first name and the last name
nameConcat = nameConcatList[0] + "." + nameConcatList[-1]

#getting the last 2 digits from the brith year
birthDigits = str(birthYear)[-2:]

#formating the email
email = nameConcat + birthDigits + "@company.com"

print(email)
```
